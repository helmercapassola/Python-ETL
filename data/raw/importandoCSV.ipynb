ANÁLISE DA NORMALIZAÇÃO

```python
#verificar ncessidade da normalização
necessidades = []
```

```python
#Verificar colunas com nomes inconsistentes 
colunas = df.columns.to_list()
if any( c != c.lower() for c in colunas):
    necessidades.append("-Colunas com maiúsculas/minúsculas inconsistentes")
    

# verificar valores categóricos insconsistentes
if 'Departament' in df.columns:
    valores = df['Department'].unique()
    if any(v != v.capitalize() for v in valores if isinstance(v , str)):
        necessidades.append("- Departamentos capitalizados inconsistentes")

#verificar espaços extras

for col in df.select_dtypes(include=['object']).columns:
    if df[col].astype(str).str.strip().ne(df[col].astype(str)).any():
        necessidades.append(f"- Espaços extras na coluna '{col}'")
        break

#Resultados
if len(necessidades) == 0:
    print("\n O dataset já está normalizado!")
    print("Não são necessárias transformações de normalização.")
    print("Os dados estão porntos para a próxima etapa do ETL.")
else:
    print("\n Foram encontradas necessidades de normalização:")
    for n in necessidades:
        print(n) 
    
    


```

```python
#Aplicar normalização
print("\n Aplicando normalização de segurança")


#Garantir que não há espaços extras

string_cols = df.select_dtypes(include=['object']).columns
for col in string_cols:
    df[col] =df[col].astype(str).str.strip()
    
    print("Normalização aplicada")
```
